# Using Label Encoding

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder = preprocessing.LabelEncoder()
Fire_Forest['month'] = LabelEncoder.fit_transform(Fire_Forest['month'])
Fire_Forest['day'] = LabelEncoder.fit_transform(Fire_Forest['day'])
Fire_Forest['size_category'] = LabelEncoder.fit_transform(Fire_Forest['size_category'])
```

```python
Fire_Forest.head()
```

# Splitting a data into features and target

```python
X = Fire_Forest.iloc[:,0:11]
y = Fire_Forest.iloc[:,-1]
```

```python
y = pd.DataFrame(y)
```

# Scaling a data using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
std_scalar =  StandardScaler()
```

```python
scaled_x = std_scalar.fit_transform(X)
```

```python
scaled_x
```

# Model Evaluation

# Tensorflow is an important library used in Neural Network

```python
!pip install tensorflow
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, Layer, Lambda
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_x,y, test_size = 0.3, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
import tensorflow as tf
model = Sequential()
model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=20, activation='tanh'))
model.add(Dense(units=20, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# Optimizers are used to avoid Vanishing/Exploding gradient in back propagation

```python
model.fit(x=X_train, y=y_train, epochs=517)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred
```

```python
FireForest_train = model.evaluate(X_train, y_train)
FireForest_train 
```

```python
FireForest_test = model.evaluate(X_test, y_test)
FireForest_test
```
